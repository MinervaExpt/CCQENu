```python
import os, sys, json

```

```python
def makejson(thedir):
    name = os.path.basename(thedir).replace("csv_","")

    print(name)
    j = {}
    j["name"] = name

    files = os.listdir(thedir)
#print (files)



    #print(keys)

    for f in files:


        if "sigma_" not in f: continue
        if "h2D" in f: continue
        #if "Q2QE" not in f: continue
        #if name not in j: j[name] = {}
        #j["name"] = name
        decode = f.split("_")
        type = decode[0]
        sample = decode[1]
        if type == "h":
            variable = decode[2]
        else:
            variable = decode[2]+"_"+decode[3]
        lines = []
        if variable not in j: j[variable] = {}    
        l = len(decode)

        observable = decode[l-3]
        #if observable != "sigma": continue

        category = decode[l-2]
        form = decode[l-1]
        o = open(os.path.join(thedir,f),'r')
        b = o.readlines()



        j[variable]["variable"] = variable                        

        lines = []
        for x in b:

            line = []
            vals = (x.strip().split(","))

            for v in vals:
                try: 

                    vi = float(v) 
                    line.append(vi)
                except:
                    line.append(v)
            #print (line)
            lines.append(line)
        o.close()
        j[variable]["type"] = type
        j[variable]["sample"] = sample
        j[variable]["variable"] = variable
        j[variable]["observable"] = observable
        j[variable]["form"] = form

        j[variable][category] = lines
        #print (j[name])

    # for k in j.keys():
    #     print ("f=",k)
    #     print(json.dumps(j[k],indent=4))
    
    f = open(name+".json",'w')
    s = json.dumps(j,indent=4) 
    f.write(s)
    f.close()
    return j

    
        
        
        
    
```
